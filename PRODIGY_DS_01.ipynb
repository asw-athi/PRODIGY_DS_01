```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
data=pd.read_csv('exercise.csv')
data.head()
```

```python
plt.bar(data['id'], data['pulse'], color = 'purple')
plt.title("Bar chart")
plt.xlabel('id') 
plt.ylabel('pulse') 
plt.show()
```

```python
plt.hist(data['pulse'], color= 'pink', edgecolor='black') 
plt.title("histogram")
plt.xlabel('Pulse')
plt.show()
```

```python
sns.swarmplot(x ='kind', y ='pulse', hue='kind', data = data)
```
