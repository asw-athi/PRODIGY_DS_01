```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
data=pd.read_csv('exercise.csv')
data.head()
```

```python
plt.bar(data['id'], data['pulse'])
plt.title("Scatter Plot")
plt.xlabel('id') 
plt.ylabel('pulse') 
plt.show()
```

```python
plt.hist(data['pulse']) 
plt.title("histogram")
plt.show()
```

```python
sns.swarmplot(x ='kind', y ='pulse', hue='kind', data = data)
```
